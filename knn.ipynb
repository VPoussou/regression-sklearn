```python
import numpy
import matplotlib.pyplot as plt

from sklearn import datasets, neighbors
```

```python
data = numpy.genfromtxt('current_data/centred_2.csv', delimiter=',', names=['x', 'y', 'z'])
print(data)
```

```python
def plot_data(data):
    plt.plot(data['x'], 'ob')
    plt.plot(data['y'], 'or')
    plt.show()
```

```python
plot_data(data)
```

```python
def plot_data_and_knn(data, k):
    problem = [data['x'], data['y']]
    solution = data['z']

    knn = neighbors.KNeighborsClassifier(k)
    knn.fit(numpy.array(problem).transpose(), solution)

    step = 0.02
    x_min, x_max = min(data['x']), max(data['x'])
    y_min, y_max = min(data['y']), max(data['y'])

    x_mesh, y_mesh = numpy.meshgrid(numpy.arange(x_min, x_max, step), numpy.arange(y_min, y_max, step))
    mesh = numpy.c_[x_mesh.ravel(), y_mesh.ravel()]
    mesh_solution = knn.predict(mesh)

    plt.figure(figsize=(5,5))
    plt.scatter(problem[0], problem[1], c=solution, cmap='coolwarm', s=20)
    plt.scatter(mesh[:,0], mesh[:,1], c=mesh_solution, cmap='coolwarm', alpha=0.1, s=20)
    plt.plot

    
```

```python
plot_data_and_knn(data, 3)
```
