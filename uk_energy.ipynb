```python
# General imports
import pandas as pd
import math
import sklearn
import numpy as np
# from statsmodels. import 
import matplotlib.pyplot as plt
```

```python
# Reading fil and checking
uk_df = pd.read_csv('energy_use_in_the_UK.csv', sep=';')
uk_df.head()
uk_df.describe()
```

```python
# Getting dates into datetime format
#and generating time index
type(uk_df['timestamp'][0])
uk_df['timestamp'] = pd.to_datetime(uk_df['timestamp'], format='mixed')
uk_df.head()
```

```python
# Let's clean up some outliers
def remove_outliers(column:pd.Series, inplace:bool) -> pd.Series:
    std = column.std()
    list_col = list(column)
    for index, el in enumerate(column):
        if el > std * 2.5:
            list_col[index] = std * 2.5
        elif el < -std * 2.5:
            list_col[index] = -std * 2.5
    if inplace:
        column = pd.Series(list_col)
    else:
        return pd.Series(list_col)

uk_df['solar'] = remove_outliers(uk_df['solar'], False)
uk_df['coal'] = remove_outliers(uk_df['coal'], False)
uk_df['nuclear'] = remove_outliers(uk_df['nuclear'], False)
uk_df['wind'] = remove_outliers(uk_df['wind'], False)
uk_df['hydro'] = remove_outliers(uk_df['hydro'], False)

uk_df.boxplot()
```
